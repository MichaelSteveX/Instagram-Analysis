```Instagram Influencer Model Prediction```

* This project focuses on building and evaluating machine learning models to predict influencer performance based on key social media metrics. The dataset contains important features such as followers, posts, average likes, engagement rate, total likes, country, and joining date, which are used to predict the target variable (Influence Score / Ranking).

* The project began with data preprocessing and feature engineering, including handling missing values, converting textual numerical formats (k, m, b) into standardized numeric values, encoding categorical variables, and transforming date features into meaningful time-based attributes.

* After data preparation, the dataset was split into training and testing sets. Multiple machine learning algorithms were implemented to compare performance, including regression models. Model evaluation was conducted using appropriate performance metrics such as R² score, Mean Absolute Error (MAE), and Root Mean Squared Error (RMSE).

* The objective of this project is to identify the most significant factors influencing influencer performance and to develop a predictive model capable of estimating influence score based on engagement and follower-related metrics.

* Overall, this project demonstrates practical application of data preprocessing, feature engineering, model building, and performance evaluation techniques in a real-world social media analytics context.

````python
# import required library
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
# import scikit learn library 
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error,r2_score
````

### 01 import dataset

````python
# Now import the data set which already cleand in previous EDA Analysis
df = pd.read_csv("C:\\Users\\mohdy\\OneDrive\\Desktop\\intern project\\top_insta_infl_cleaned_data.csv")
````

````python
# copy dataset 
data = df.copy()
````

````python
# print 
data.head()
````

````python
data.columns
````

### 02 Feature selection

````python
# Feature selection
X = data[['followers', 'avg_likes', 'engagement_rate', 'new_post_avg_like', 'like_follower_ratio', 'post_follower_ratio']]
y = data['influence_score']
````

### 03 Train_Test_Split

````python
x_train,x_test,y_train,y_test = train_test_split(X, y, test_size=0.2, random_state=52)
````

### 04 Standardize features

````python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(x_train)
X_test_scaled = scaler.transform(x_test)
````

### 05 Model Apply

````python
# Apply linear Regression model 
rf = RandomForestRegressor(n_estimators=30)
rf.fit(x_train, y_train)

# This is for Score
rf.score(x_test, y_test)*100 , rf.score(x_train, y_train)*100
````

### 06 Model Predictions and evaluation

````python
y_pred = rf.predict(X_test_scaled)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
print(f'Mean Squared Error: {mse}')
print(f'R^2 Score: {r2}')
````
